## 4.2: Deformation of an axial rod

- Description: 
    - 120-inch-long steel pole supporting a floor system
    - Post-construction settling: 
        - Top: 0.5 inch
        - Bottom: 1.0 inch
    - What is the tensile stress that this settling has now introduced into the system? (Recall that stress is proportional to strain, which is the derivative of the displacement.)
- Domain 
    - $\Omega = [0,1]$
- DE: 
    - $EA u_{xx} + f = 0$ on $\Omega$
- Constants: 
    - $E = 30,000 \frac{kip}{in^2}$ (modulus of elasticity)
    - $r = 2.0  in$ (radius of pole/bar)
    - $A = \pi r^2 \,in^2$  (cross-sectional area of the bar)
    - $f(x) = 0$  (rightward force per unit length) 
- BCs
    - Dirichlet: $u(0) = g_0 = 0.5 in$
    - Dirichlet: $u(L) = g_L = 1.0 in $
- Unknowns/QOIs: 
    - $u(x)$ = displacement as a function of $x$
    - Derivative of displacement (strain) . . . somewhere?
    - Proportionality constant for stress --> strain ?


```python
E = 3.0e4
r = 2.0
A = np.pi * r**2.0
EA = E * A
f = lambda x: 0.0
g_0 = -0.5
g_L = -1.0
```

```python

bcl = bc.BoundaryCondition("left", "Dir", 1, 0, g_0)
bcr = bc.BoundaryCondition("right", "Dir", 1, 0, g_L)

f = lambda x: 0

p1 = fp.Problem(bcl, bcr, f, 10)
```

```python
s = p1.SolveOnly()
```
